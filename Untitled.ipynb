```python
def clean_url_array (array):
    res = []
    for url in array:
        print(type(url))
        print(url)
        valid = validators.url(url)
        if valid == True:
            res.append(url)
        else:
            continue
    return res
```

```python
def array_to_folder (array, name):
    os.mkdir(name)
    for i, url in enumerate(array):
        img_id = str(i)
        urllib.urlretrieve(url, f"{name}/{img_id}.jpg")
```
